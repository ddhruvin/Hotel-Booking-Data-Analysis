```python
df['adr'].plot(kind='line')
#as seen, only one value is a spike in the plot => remove it
```

```python
#to see amount of cancellations and amount of non-cancellations in the Dataset in terms of %
#normalize=True returns % value
#63% bookings are not calceled
#37% bookings are canceled which is very high(5%-10% is manageable)
canceled_perc = df['is_canceled'].value_counts(normalize=True)
print(canceled_perc)

plt.figure(figsize=(5,4))
plt.title('Reservation Status')
plt.bar(['Not Calcelled','Calcelled'], df['is_canceled'].value_counts(), edgecolor='black', width=0.7)
plt.show()
```

```python
#as seen, calceled is almost as half as not-calceled
#hue is for defining the basis of count (here, is_canceled)
plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel',hue='is_canceled', data=df, palette='Reds')
plt.title("Reservation status in different hotels", size=20)
plt.xlabel("Hotels")
plt.ylabel("Number of Reservations")
plt.show()
#ratio of not-canceled to canceled is more in cities than in resort hotels, so find reasons to that
```

```python
df.columns
#just to see name of columns
```

```python
resort_hotels = df[df['hotel'] == 'Resort Hotel']
resort_hotels['is_canceled'].value_counts(normalize=True)
#to check resort hotels more deeply
#as seen, 28% bookings are canceled
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
#as seen, 42% bookings are canceled (very high)
```

```python
#now check whether price is affecting bookings in cuty hotels?
resort_hotels=resort_hotels.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(8,4))
plt.title("Average Daily Rate in City and Resort Hotels", size=20)
plt.plot(resort_hotels.index, resort_hotels['adr'], label='Resort Hotels')
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotels')
plt.legend(fontsize=10, loc='lower right')
plt.show()
#there are certain spikes in the plot, why are they? We can consider that they are because of weekends
```

```python
#now see which month has highest reservations and which month has most cancellations
#we'll use countplot so that we can compare
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month',hue='is_canceled', data=df, palette='Reds')
plt.title("Reservation status per month", size=20)
plt.xlabel("Months")
plt.ylabel("Number of Reservations")
plt.legend(['Not Cancelled','Cancelled'], loc='upper right', fontsize=10)
plt.show()
#as seen, january has most no. of cancellations and august has the lowest
#insight => when there are more reservations, cancellations are low and vice versa, so look adr acc. to month
```

```python
plt.figure(figsize=(15,8))
plt.title("ADR per Month", size=30)
sns.barplot(x='month', y='adr', data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
#in aug, adr is lowest and correspondingly no. of cancellations and august is the lowest
#in jan, adr is highst and correspondingly no. of cancellations and january is the highest
#hypothesis proved
```

```python
canceled_df_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace = True)
canceled_df_adr.sort_values('reservation_status_date', inplace = True)

not_canceled_data = df[df['is_canceled'] == 0]
not_canceled_df_adr = not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_df_adr.reset_index(inplace = True)
not_canceled_df_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize =(20,6))
plt.title('Average Daily Rate')
plt.plot(not_canceled_df_adr['reservation_status_date'],not_canceled_df_adr['adr'], label = 'Not cancelled')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'], label = 'Cancelled')
plt.show()

#as seen this is a messy data
#we have few values before 2015 and after 2017
#so we'll focus on the data we have and neglect the rest of the data.
```

```python
canceled_df_adr = canceled_df_adr[(canceled_df_adr['reservation_status_date']>'2016') & (canceled_df_adr['reservation_status_date']<'2017-09')]
not_canceled_df_adr = not_canceled_df_adr[(not_canceled_df_adr['reservation_status_date']>'2016') & (not_canceled_df_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize =(20,6))
plt.title('Average Daily Rate')
plt.plot(not_canceled_df_adr['reservation_status_date'],not_canceled_df_adr['adr'], label = 'Not cancelled')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'], label = 'Cancelled')
plt.legend(fontsize=20)
plt.show()

#as seen price has a direct effect on cancellations
```
